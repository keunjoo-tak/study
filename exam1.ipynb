# 작업형 1
mtcars 데이터셋 (data/mtcars.csv)의 qsec컬럼을 최소최대 척도(Min-Max Scale)로 변환한 후 0.5보다 큰 값을 가지는 레코드 수를 구하시오


* dataset 위치 : data/mtcars.csv

```python
import pandas as pd
df = pd.read_csv('mtcars.csv')
print(df)
```

```python
qsec = df.loc[:, 'qsec']
print(qsec)
qsec_min, qsec_max = qsec.min(), qsec.max()
res = (qsec-qsec_min)/(qsec_max-qsec_min)
print(res)
```

```python
a = res >= 0.5
print(a)
b = df[a]
print(b)

print('record: ', len(b))
```
